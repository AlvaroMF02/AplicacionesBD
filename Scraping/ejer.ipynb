```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
header={
    "User-Agent":"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36",
}

r = requests.get('https://meneame.net/', headers=header)
        # como si fueras uno nuevo cada rato
r.status_code
```

```python
# sequir co la misma sesio
sesion = requests.session()
response = sesion.get('https://meneame.net/', headers=header)
response.status_code()
```

```python
soup = BeautifulSoup(response.text,"html.parser")
```

```python
soup
```

```python
noticias = soup.find_all("div",class_="news-summary")
# noticias[0]
noticias
```

```python
votos = soup.find_all("div",class_="votes")
votos
```

```python
for x in noticias:
    meneo = re.sub(r'\D+', '', x.find("div",class_="votes").get_text(strip=True))
    clics = re.sub(r'\D+', '', x.find("div",class_="clics").get_text(strip=True))
    fuente = x.find("span",class_="showmytitle").get_text(strip=True)
    comentarios = re.sub(r'\D+', '', x.find("a",class_="comments ga-event").get_text(strip=True))
    print(f"Meneos = {meneo}, clics = {clics}, fuente = {fuente}, comentarios = {comentarios}")
```
